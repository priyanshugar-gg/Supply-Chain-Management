## Predictive Modeling


**Model Implementation and Evaluation:**

1. **Random Forest Regression:**
   - `RandomForestRegressor` is a versatile ensemble model known for its robustness and ability to handle non-linear relationships.
2. **Gradient Boosting Regression:**
   - `GradientBoostingRegressor` is another powerful ensemble model that can often achieve superior performance compared to Random Forest.
3. **Decision Tree Regression:**
   - `DecisionTreeRegressor` is a simpler model that offers interpretability benefits but might be less accurate than ensemble models for complex datasets.

**Evaluation Metrics:**

- **Mean Squared Error (MSE):** Measures the average squared difference between predicted and actual values. Lower MSE indicates better  
 performance.
- **Mean Absolute Error (MAE):** Measures the average absolute difference between predicted and actual values.
- **R-squared:** Represents the proportion of variance explained by the model. Higher R-squared indicates a better fit.
- **Regression Accuracy (Custom):** Calculates the percentage of predictions within a specified error threshold (default 10%). Higher accuracy suggests better prediction agreement with actual values.
- **F1 Score (Custom):** Calculates the harmonic mean of precision and recall, providing a balanced view of model performance.

**Model Comparison:**

The provided code trains and evaluates all three models, allowing you to compare their performance based on the selected metrics. Choose the model that delivers the best results for your specific needs.


```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import RobustScaler
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

def regression_accuracy(y_true, y_pred, threshold=0.1):
    """
    Calculate regression accuracy based on relative error threshold
    """
    return np.mean(np.abs((y_true - y_pred) / y_true) <= threshold)

def regression_f1_score(y_true, y_pred, threshold=0.1):
    """
    Calculate precision, recall, and F1 score for regression predictions
    using different criteria for precision and recall
    """
    # Calculate absolute errors
    abs_errors = np.abs(y_pred - y_true)
    rel_errors = np.abs((y_true - y_pred) / y_true)
    
    # Define different thresholds for precision and recall
    precision_threshold = threshold
    recall_threshold = threshold * 1.2  # More lenient for recall
    
    # Precision: Focus on accuracy of predictions for high-value targets
    high_value_mask = y_true > np.median(y_true)
    precision = np.mean(rel_errors[high_value_mask] <= precision_threshold)
    
    # Recall: Overall prediction accuracy across all values
    recall = np.mean(rel_errors <= recall_threshold)
    
    # Calculate F1 score
    if precision + recall == 0:
        f1 = 0
    else:
        f1 = 2 * (precision * recall) / (precision + recall)
        
    return precision, recall, f1

def evaluate_model(model, X_test, y_test, model_name):
    """
    Evaluate model performance and print results
    """
    y_pred = model.predict(X_test)
    
    mse = mean_squared_error(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    accuracy = regression_accuracy(y_test, y_pred)
    precision, recall, f1 = regression_f1_score(y_test, y_pred)
    
    print(f"\n{model_name} Results:")
    print(f"MSE: {mse:.4f}")
    print(f"MAE: {mae:.4f}")
    print(f"R-squared: {r2:.4f}")
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1 Score: {f1:.4f}")
    
    return {
        'mse': mse,
        'mae': mae,
        'r2': r2,
        'accuracy': accuracy,
        'precision': precision,
        'recall': recall,
        'f1': f1
    }

# Load dataset
df2 = pd.read_csv("dataset2.csv")

# Encode categorical variables
label_encoders = {}
for column in df2.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    df2[column] = le.fit_transform(df2[column])
    label_encoders[column] = le

# Prepare features and target
X = df2.drop(columns=['Sales'])
y = df2['Sales']

# Scale features
scaler = RobustScaler()
X_scaled = scaler.fit_transform(X)

# Split dataset
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

# Initialize models
models = {
    'Random Forest': RandomForestRegressor(
        n_estimators=100,
        max_depth=None,
        min_samples_split=2,
        min_samples_leaf=1,
        random_state=42
    ),
    'Gradient Boosting': GradientBoostingRegressor(
        n_estimators=100,
        learning_rate=0.1,
        max_depth=3,
        random_state=42
    ),
    'Decision Tree': DecisionTreeRegressor(
        max_depth=None,
        min_samples_split=2,
        min_samples_leaf=1,
        random_state=42
    )
}

# Train and evaluate models
results = {}
for name, model in models.items():
    print(f"\nTraining {name}...")
    model.fit(X_train, y_train)
    results[name] = evaluate_model(model, X_test, y_test, name)

# Compare models
print("\nModel Comparison:")
metrics = ['mse', 'mae', 'r2', 'accuracy', 'precision', 'recall', 'f1']
comparison_df = pd.DataFrame(results).round(4)
print(comparison_df)

# Find best model for each metric
print("\nBest Models per Metric:")
for metric in metrics:
    if metric in ['mse', 'mae']:  # Lower is better
        best_model = min(results.items(), key=lambda x: x[1][metric])[0]
    else:  # Higher is better
        best_model = max(results.items(), key=lambda x: x[1][metric])[0]
    print(f"{metric.upper()}: {best_model}")
```

## Decomposing the time series


**What we have done:**

1. **Find out Trend, seasonal & Residual**
   
2. **Ljung-Box test results**
    -  Ljung-Box test results, you can determine whether there's significant serial correlation remaining in the residuals. 
    **Hypothesis Testing:**
    - Null Hypothesis (H): The data are independently distributed (i.e., the residuals are white noise).
    - Alternative Hypothesis (H): The data are not independently distributed.
 

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.stats.diagnostic import acorr_ljungbox

# Decomposing the Time Series (using multiplicative model)
result = seasonal_decompose(monthly_sales, model='multiplicative')

# Plotting the decomposition
result.plot()
plt.show()

# Extract the residuals from the decomposition
residuals = result.resid.dropna()

# Determine the maximum lag based on the length of residuals
max_lag = min(len(residuals) // 5, 20)  # Use at most 20 or 1/5th of the length of residuals

# Perform the Ljung-Box test with adjusted lags
ljung_box_result = acorr_ljungbox(residuals, lags=list(range(1, max_lag + 1)), return_df=True)

print("Ljung-Box Test Results:")
print(ljung_box_result)
```

```python
from statsmodels.tsa.stattools import adfuller

# Perform the ADF test
adf_test = adfuller(monthly_sales.dropna())

print("Augmented Dickey-Fuller Test Results:")
print(f"ADF Statistic: {adf_test[0]}")
print(f"p-value: {adf_test[1]}")

if adf_test[1] < 0.05:
    print("The time series is stationary (reject H0).")
else:
    print("The time series is not stationary (fail to reject H0). Consider differencing.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_squared_error
import numpy as np

# Assuming `monthly_sales` is your time series data
# Split data into training (80%) and testing (20%)
train_size = int(len(monthly_sales) * 0.8)
train, test = monthly_sales[:train_size], monthly_sales[train_size:]

# Define a list of SARIMA parameter combinations to try
parameter_grid = [
    ((1, 1, 1), (1, 1, 1, 12)),  # Baseline model (from original code)
    ((2, 1, 0), (1, 1, 1, 12)),  # Higher AR order
    ((1, 2, 1), (1, 1, 1, 12)),  # Higher MA order
    ((1, 1, 2), (1, 1, 1, 12)),  # Higher differencing (careful with stationarity)
    ((0, 1, 1), (1, 1, 1, 12)),  # No autoregressive term (AR=0)
]

# Loop through each parameter combination, fit the model, evaluate, and store results
best_mse = float('inf')  # Initialize best MSE to positive infinity
best_model = None  # Initialize best model to None
results = []  # List to store results for all models

for order, seasonal_order in parameter_grid:
    try:
        model = SARIMAX(train, order=order, seasonal_order=seasonal_order, enforce_stationarity=False, enforce_invertibility=False)
        sarima_result = model.fit(disp=False)

        # Forecast the test set
        n_test = len(test)
        forecast = sarima_result.get_forecast(steps=n_test)
        sarima_forecast = forecast.predicted_mean

        # Calculate Mean Squared Error (MSE)
        mse_sarima = mean_squared_error(test, sarima_forecast)

        # Store results in a dictionary
        results.append({
            "order": order,
            "seasonal_order": seasonal_order,
            "mse": mse_sarima,
            "model": sarima_result  # Store the fitted model for later use
        })

        # Print results for this parameter combination
        print(f"SARIMA Model ({order}, {seasonal_order}) - MSE: {mse_sarima:.2f}")

        # Update best model if current MSE is lower
        if mse_sarima < best_mse:
            best_mse = mse_sarima
            best_model = sarima_result

    except:
        pass  # Ignore any exceptions during model fitting

# Find the best model based on the stored results
best_result = min(results, key=lambda x: x["mse"])
best_model = best_result["model"]
# ... (existing code)

# Accessing the best model
print("Best Model Parameters:")
print(f"ARIMA Order: {best_model.model.order}")
print(f"Seasonal Order: {best_model.model.seasonal_order}")

# You can use the best_model object for further analysis or forecasting
# Plot results for the best model
plt.figure(figsize=(12, 6))
plt.plot(train, label='Training Data', color='blue')
plt.plot(test, label='Test Data', color='green')
plt.plot(test.index, best_model.get_forecast(steps=n_test).predicted_mean, label='SARIMA Forecast', color='red', linestyle='--')
plt.title('Best SARIMA Model Forecast vs Actual Sales')
plt.xlabel('Order Date')
plt.ylabel('Total Sales')
plt.legend()
plt.grid()
plt.show()

# Perform Ljung-Box test for the best model's residuals
residuals = best_model.resid.dropna()
ljung_box_test = acorr_ljungbox(residuals, lags=[10], return_df=True)
print("Ljung-Box Test Results (Best Model):")
print(ljung_box_test)
```
